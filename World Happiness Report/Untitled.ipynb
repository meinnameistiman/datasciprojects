Dünya Mutluluk Raporu Projesi

Bu veri seti ilk kez 2012 yılında yayınlanmıştır ve 155 ülkeyi mutluluk seviyelerine göre sıralayan dünya mutluluk raporlarından oluşmaktadır yapılan bu çalışmanın amacı ise yıllara göre dünya mutluluk oranlarını incelemek, analiz yapmak ve verileri görselleştirmektir. Mutluluk seviyelerini belirlemek için 6 faktör -ekonomik üretim, sosyal destek, yaşam beklentisi, özgürlük, yolsuzluğun olmaması ve cömertlik - göz önüne alınmaktadır. Amaç günümüz dünyasında daha mutlu toplumlar oluşmasına katkı sağlamak için aşağıdaki gibi sorulara cevaplar arayabilmektir.

Projede yer alan görevler: 
* Hangi ülkeler veya bölgeler genel mutlulukta ve mutluluğa katkıda bulunan       altı faktörün her birinde en yüksek sırada yer alıyor?
* 2015-2016 ile 2016-2017 raporları arasında ülke sıralamaları veya puanları     nasıl değişti?
* Herhangi bir ülke mutlulukta önemli bir artış veya azalma yaşadı mı?

Detay ve kaynak için : https://www.kaggle.com/datasets/unsdsn/world-happiness

```python
#kütüphanelerin eklenmesi

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#verinin proje icine cagirilmasi ve goruntulenmesi

data2015 = pd.read_csv("archive/2015.csv")
data2016 = pd.read_csv("archive/2016.csv")
data2017 = pd.read_csv("archive/2017.csv")
data2018 = pd.read_csv("archive/2018.csv")
data2019 = pd.read_csv("archive/2019.csv")
```

## 2015 Mutluluk oranları

```python
data2015.columns
```

```python
numeric_features = ["Economy (GDP per Capita)", "Family",
                    "Health (Life Expectancy)", "Freedom", "Trust (Government Corruption)",
                   "Generosity"]

for feature in numeric_features:
    plt.figure(figsize=(6, 1))
    
    plt.subplot()
    plt.title(f"{feature} - Top 10 Happy Country")
    sns.barplot(data=data2015.head(10), x="Country", y=feature)
    plt.grid(False)
    plt.xticks(rotation=60)
    plt.ylabel(None)
    plt.xlabel(None)

    
    plt.show()
```
